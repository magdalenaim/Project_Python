2. SKILL DEMAND

What are the most demanded skills for the top 3 most popular data roles?

Methodology
- Clean-up skill column
- Calculate skill count based on job_title_short
- Calculate skill percentage
- Plot final findings

Filter Data for Poland and Czechia

```python
df_CZPL = df[(df['job_country'] == 'Poland')|(df['job_country'] == 'Czechia')]
```

Explode Skills

```python
df_skills = df_CZPL.explode('job_skills')

df_skills[['job_title', 'job_skills']]
```

Count Skills per Job Title

```python
# Group by job_skills and job_title_short and count the number of occurrences
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short', 'job_country']).size()

# Name the count column as count
df_skills_count = df_skills_count.reset_index(name='skill_count')

# Sort the values by skill_count in descending order
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

Create List of Top 3 Roles
Focus: Data Analysts, Data Engineers, and Data Scientists

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles = sorted(job_titles[:3])

job_titles
```

Plot Skill Counts

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming job_titles is a predefined list of job titles
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']  # Replace with your actual list of job titles
countries = ['Poland', 'Czechia']

# Set the theme for seaborn
sns.set_theme(style='ticks')

# Create the subplots, 2 rows (one for each country) and N columns (one for each job title)
fig, ax = plt.subplots(len(countries), len(job_titles), figsize=(15, 10))

for country_idx, country in enumerate(countries):
    for job_idx, job_title in enumerate(job_titles):
        df_plot = df_skills_count[(df_skills_count['job_title_short'] == job_title) & 
                                  (df_skills_count['job_country'] == country)].head(5)[::-1]
        sns.barplot(data=df_plot, x='skill_count', y='job_skills', ax=ax[country_idx, job_idx], hue='skill_count', palette='dark:b_r')
        ax[country_idx, job_idx].set_title(f'{job_title} in {country}')
        ax[country_idx, job_idx].invert_yaxis()
        ax[country_idx, job_idx].set_ylabel('')
        ax[country_idx, job_idx].set_xlabel('')
        ax[country_idx, job_idx].get_legend().remove()
        ax[country_idx, job_idx].set_xlim(0, 2500)  # make the scales the same


# Set the overall title and adjust the layout
fig.suptitle('Counts of Skills Requested in Poland and Czechia Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5, w_pad=0.5)  # fix the overlap
plt.show()
```

Convert Counts to Percentages
Focus: Counts don't show what portions of jobs are requesting these skills

```python
# Use original df to get the count of job titles
df_job_title_count = df_CZPL['job_title_short'].value_counts().reset_index(name='jobs_total')

df_job_title_count
```

```python
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, on='job_title_short', how='left')

df_skills_perc['skill_percent'] = (df_skills_perc['skill_count'] / df_skills_perc['jobs_total']) * 100

df_skills_perc
```

Plot Percentage Count

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter

# Assuming job_titles is a predefined list of job titles
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']  # Replace with your actual list of job titles
countries = ['Poland', 'Czechia']

# Set the theme for seaborn
sns.set_theme(style='ticks')

# Create the subplots, 2 rows (one for each country) and N columns (one for each job title)
fig, ax = plt.subplots(len(countries), len(job_titles), figsize=(15, 10))

for country_idx, country in enumerate(countries):
    for job_idx, job_title in enumerate(job_titles):
        df_plot = df_skills_count[(df_skills_count['job_title_short'] == job_title) & 
                                  (df_skills_count['job_country'] == country)].head(5)[::-1]
        total = df_plot['skill_count'].sum()
        df_plot['skill_count'] = df_plot['skill_count'] / total * 100  # Convert to percentage
        
        sns.barplot(data=df_plot, x='skill_count', y='job_skills', ax=ax[country_idx, job_idx], hue='skill_count', palette='dark:b_r')
        ax[country_idx, job_idx].set_title(f'{job_title} in {country}')
        ax[country_idx, job_idx].invert_yaxis()
        ax[country_idx, job_idx].set_ylabel('')
        ax[country_idx, job_idx].set_xlabel('')
        ax[country_idx, job_idx].get_legend().remove()
        ax[country_idx, job_idx].set_xlim(0, 37)  # Percentage scale
    
        # Format the x-axis to show percentages
        ax[country_idx, job_idx].xaxis.set_major_formatter(PercentFormatter(decimals=0))

# Set the overall title and adjust the layout
fig.suptitle('Likelihood of Skills Requested in Poland and Czechia Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5, w_pad=0.5)  # fix the overlap
plt.show()
```

3. SKILL TREND

HOW ARE IN-DEMAND SKILLS TRENDING FOR DATA ANALYSTS?

Methology
1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

Group the Skills by Month
- Select only those job postings that are for Data Analysts and the job country is Poland and Czechia. Then extract the month from each job's posting date to see when jobs are listed. Next, converts a column that lists skills into a usable list format. Finally, it rearranges the data so that each skill from the list gets its own row.

```python
## Skill count per month for data analysts
df_DA_PLCZ = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'Poland')|(df['job_country'] == 'Czechia')].copy()

df_DA_PLCZ['job_posted_month_no'] = df_DA_PLCZ['job_posted_date'].dt.month

# Explode job_skills column
df_DA_PLCZ_explode = df_DA_PLCZ.explode('job_skills')
```

Pivot in Prep for Plotting
- Create a pivot table from the df_DA_PLCZ_explode, setting 'month' as the index, job_skills as the columns, and fills missing entries with zero.

```python
# Group by month and job_skills
df_DA_PLCZ_pivot = df_DA_PLCZ_explode.pivot_table(index='job_posted_month_no', columns='job_skills',  aggfunc='size', fill_value=0)

df_DA_PLCZ_pivot
```

Sort columns by count and change month numbers to names
- It adds a new row labeled Total that sums up counts across all months for each skill. Finally, it reorders the columns based on the total counts, displaying them from highest to lowest, and shows the updated pivot table.

```python
# sorts by count
df_DA_PLCZ_pivot.loc['Total'] = df_DA_PLCZ_pivot.sum()
df_DA_PLCZ_pivot = df_DA_PLCZ_pivot[df_DA_PLCZ_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_PLCZ_pivot = df_DA_PLCZ_pivot.drop('Total')

df_DA_PLCZ_pivot
```

Plot the Monthly Skill Counts
- Plot a line chart of the top 5 skills for data analysts, indexed by month. It selects the first five columns and plots them.

```python
df_DA_PLCZ_pivot.iloc[:, :5].plot(kind='line')

plt.title('Trending Top Skills for Data Analysts in Poland and Czechia')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

This is good but totals can be misleading without context. Percentages clarify relationships within the whole, giving a more accurate comparison.

Calculate Percentage of Total Jobs
- Calculates the total number of entries for each month in df_DA by grouping the data by 'month' and counting the number of occurrences in each group, storing the results in DA_totals. This is needed to compute the percentages.

```python
# Get monthly totals
DA_totals = df_DA_PLCZ.groupby('job_posted_month_no').size()

DA_totals
```

This code calculates the percentage of each skill's occurrence for each month by dividing the first 12 rows of the df_DA_pivot DataFrame, which contains skill counts, by the monthly totals in DA_totals and multiplying by 100. The new DataFrame, df_DA_pivot_percent, shows skill distributions as percentages of total monthly job entries.

```python
# divide first 12 rows of df_DA_pivot by DA_totals
df_DA_PLCZ_percent = df_DA_PLCZ_pivot.iloc[:12].div(DA_totals/100, axis=0)

# changes month number to month name
df_DA_PLCZ_percent = df_DA_PLCZ_percent.reset_index()
df_DA_PLCZ_percent['job_posted_month'] = df_DA_PLCZ_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_PLCZ_percent = df_DA_PLCZ_percent.set_index('job_posted_month')
df_DA_PLCZ_percent = df_DA_PLCZ_percent.drop(columns='job_posted_month_no')

df_DA_PLCZ_percent
```

Plot Monthly Skill Demand
- Creates a line plot for the top five skills of data analysts, shown as percentages of the total job entries per month, using the first 5 columns of the df_DA_pivot_percent DataFrame. Also the legend is moved outside of the plot for readability.

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_PLCZ_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in Poland and Czechia')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```

Coloring by Technology

- Removes duplicate entries and any rows with missing values. Then, it combines all dictionary entries into a single dictionary, summing values for keys that repeat across entries. To ensure each key's values are unique, it converts the values to a set and then back to a list. The final result is a dictionary where each key represents a skill and each value is a list of unique attributes associated with that skill.

```python
df_technology = df['job_type_skills'].copy()

# remove duplicates
df_technology = df_technology.drop_duplicates()

# remove NaN values
df_technology = df_technology.dropna()

# combine all dictionaries into one
technology_dict = {}
for row in df_technology:
    row_dict = ast.literal_eval(row)  # convert string to dictionary
    for key, value in row_dict.items():
        if key in technology_dict:  # if key already exists in technology_dict, add value to existing value
            technology_dict[key] += value
        else:                       # if key does not exist in technology_dict, add key and value
            technology_dict[key] = value

# remove duplicates by converting values to set then back to list
for key, value in technology_dict.items():
    technology_dict[key] = list(set(value))

technology_dict
```

- This code converts a dictionary into a pandas DataFrame, then explodes the 'skills' column, creating multiple rows for each 'technology' entry if it contains multiple skills.

```python
# turn dictionary into dataframe
df_technology = pd.DataFrame(list(technology_dict.items()), columns=['technology', 'skills'])

df_technology = df_technology.explode('skills')

df_technology
```
